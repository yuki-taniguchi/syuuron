# 分析結果を集計するノートブック
- 一旦AUCだけで比較してみる
- output/ファイルから各手法の各データの結果をまとめる
- それらの平均をまとめる
- 分散分析とかグラフとかにして表示する

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
multi_columns = [
    np.array(["MTS", "MTS", "MTSBag", "MTSBag", "SMOTEMTSBag", "SMOTEMTSBag"]),
    np.array(["mean", "std", "mean", "std", "mean", "std"]),
]
method = ['MTS', 'MTSBag', 'SMOTEMTSBag']
data = ['abalone', 'cancer', 'car', 'letter', 'wine', 'yeast']
e_index = ['AUC', 'accuracy', 'recall', 'Specificity', 'precision', 'gmeans', 'RS']
```

```python
index_result_df.loc[data[0], method[0]]['mean'] = 1
```

```python
index_result_df = pd.DataFrame(
                index=data,
                columns=multi_columns,
                )
index_result_df[method[0]]['mean'][data[0]] = 1
index_result_df
```

```python
for e in e_index:
    index_result_df = pd.DataFrame(
                index=data,
                columns=multi_columns,
                )
    for m in method:
        for d in data:     
            df = pd.read_csv(f'../data/output/{m}_{d}_result.csv', index_col=0)
            index_result_df.loc[d, m]['mean'] = df[e].mean()
            index_result_df.loc[d, m]['std'] = df[e].std()
    print(e)
    display(index_result_df)
```
